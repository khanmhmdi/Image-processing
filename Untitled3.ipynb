```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

from PIL import Image
from numpy import asarray
from sklearn.cluster import KMeans
```

```python
def img_color_clustering(n_cluster):
    img =  np.array(Image.open('/home/mkhanmhmdi/Downloads/image2.jpg'))
    img = asarray(img)
    img=img.reshape((img.shape[1]*img.shape[0],3))
    img.setflags(write=1)
    kmeans = KMeans(n_clusters=n_cluster).fit(img)
    labels=kmeans.labels_
    print(labels)
    labels=list(labels)
    centroid=kmeans.cluster_centers_
    print(centroid)
    
    percent=[]
    for i in range(len(centroid)):
      j=labels.count(i)
      j=j/(len(labels))
      percent.append(j)
    print(percent)
    plt.pie(percent,colors=np.array(centroid/255),labels=np.arange(len(centroid)))
    plt.show()
    
    indx = pd.DataFrame(kmeans.labels_,columns=['l'])
    for i in range(n_cluster):
        img[indx.loc[indx['l']==i].index] = centroid[i]
    img = img.reshape((480, 640, 3))
    img = Image.fromarray(img)
    # Display the Numpy array as Image
    img.show()
    img.save("img1_{}_colors.jpeg".format(n_cluster))
    return np.array(img)
```

```python
img_list = []
for i in range(1,20):
    img_list.append(img_color_clustering(i))
```

```python
def image_indexing(img):
    unique_image_values = np.unique(img)    
    indexed_image = {}
    indexed_image['image_shape'] = img.shape
    for i in unique_image_values:
        indexed_image[i] = np.where(img==0)
    return indexed_image
```

```python
import cv2
import sys

# imagePath = sys.argv[1]

image = cv2.imread('/home/mkhanmhmdi/Downloads/image1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faceCascade = cv2.CascadeClassifier("/home/mkhanmhmdi/Downloads/haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.2,
    minNeighbors=5,
    minSize=(100,100)
)

print("[INFO] Found {0} Faces!".format(len(faces)))

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

status = cv2.imwrite('faces_detected.jpg', image)
print("[INFO] Image faces_detected.jpg written to filesystem: ", status)
```

```python
import cv2
import numpy as np

def warpAffine(img, M, dsize):
    output = np.zeros(dsize, dtype=np.uint8)
    for y in range(dsize[0]):
        for x in range(dsize[1]):
            p = np.dot(np.linalg.inv(M), np.array([x, y, 1]))
            if p[0] >= 0 and p[1] >= 0 and p[0] < img.shape[1] and p[1] < img.shape[0]:
                
                fx = p[0] - int(p[0])
                fy = p[1] - int(p[1])
                x1 = int(p[0])
                x2 = min(x1 + 1, img.shape[1] - 1)
                y1 = int(p[1])
                y2 = min(y1 + 1, img.shape[0] - 1)
                output[y, x] = (1 - fx) * (1 - fy) * img[y1, x1] + fx * (1 - fy) * img[y1, x2] + (1 - fx) * fy * img[y2, x1] + fx * fy * img[y2, x2]
    return output
```

```python
for i in range(3):
    
    shear_factor = 0.5
    shear_matrix = np.float32([[1, shear_factor, 0], [0, 1, 0]])
    shear_width = int(l[i].shape[1] + abs(shear_factor * l[i].shape[0]))
    shear_height = l[i].shape[0]
    img_shearing = cv2.warpAffine(l[i], shear_matrix, (shear_width, shear_height))
    
    cv2.imshow('Shearing', img_shearing)
    cv2.waitKey(0)
```

# Blur transformation

```python
# Apply a Gaussian blur with kernel size (5,5) and sigmaX = 0
blurred = cv2.blur(img, (5,5), 0)

# Display the original a/nd blurred images
cv2.imshow('Original Image', img)
cv2.imshow('Blurred Image', blurred)

```

```python
import cv2
import numpy as np
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size**2)

pad_size = kernel_size // 2
padded_img = cv2.copyMakeBorder(img, pad_size, pad_size, pad_size, pad_size, cv2.BORDER_CONSTANT, value=0)

convolved = np.zeros_like(img, dtype=np.float32)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            convolved[i,j,k] = np.sum(kernel * padded_img[i:i+kernel_size, j:j+kernel_size, k])

blurred = np.uint8(convolved)
cv2.imshow('Original Image', img)
cv2.imshow('Blurred Image', blurred)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
